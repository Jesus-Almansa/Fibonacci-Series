```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rango = 50
def fibonacci(value):
    lista = []
    c = 1
    for n in range(0,value):
        lista.append(c)
        if len(lista) > 1:
            c = lista[n] + lista[n-1]
    return lista
```

```python
def fibonacci2(value):
    lista = []
    a, b = 0, 1
    for i in range(value):
        a, b = b, a + b
        lista.append(a)
    return lista
```

```python
print(fibonacci(rango))
```

```python
print(fibonacci2(rango))
```
